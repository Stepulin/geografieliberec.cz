```python
isocodes = pd.read_csv("iso_codes-countries.csv", sep=";", encoding="utf-8")
latlng = pd.read_csv("lat_lng-iso_2.csv", sep=";", encoding="utf-8")
```

```python
isolatlng = pd.merge(isocodes, latlng, on="id")
```

```python
print(isolatlng)
```

```python
isolatlng.rename(columns={"iso_3_x":"iso_3"}, inplace=True)
isolatlng = isolatlng.drop(["iso_3_y"], axis=1)
isolatlng.to_csv("isolatlng_and_name.csv", encoding="utf-8", index=False)
```

```python
datamerge = pd.merge(databeta2, isolatlng, left_on="LocID", right_on="id")
```

```python
# print(datamerge)
# datamerge.head()
# datamerge.info()
```

```python
extradata = pd.read_csv("indikatory_pomocne.csv", sep=";", encoding="utf-8")
datamergetemp = pd.merge(isolatlng, extradata, left_on="id", right_on="LocID")
datamergetemp = datamergetemp.drop(["id","iso_2","iso_3","country_en","country_cz","latitude","longitude","Location","MidPeriod","Variant","Time","VarID"], axis=1)
```

```python
extradata.info()
datamergetemp.info()
```

```python
#datafinal = datamerge.drop(["LocID", "iso_3_y","Location"], axis=1)
datafinal2 = datamerge.drop(["Location","TFR","NRR","Births","IMR","Q5","Deaths","DeathsMale","DeathsFemale","SRB","MAC"], axis=1)
datafinal2.rename(columns={"iso_3_x":"iso_3","LocID":"loc_id","CBR":"porodnost","LEx":"str_delka_ziv-CLKM","LExMale":"str_delka_ziv-M","LExFemale":"str_delka_ziv-Z","CDR":"umrtnost","CNMR":"mig_saldo_na_1000","NetMigrations":"mig_saldo","GrowthRate":"procentualni_narust_pop","NatIncr":"prirozeny_prirustek_na_1000","Time":"time","MidPeriod":"midperiod"}, inplace=True)
datafinal2.to_csv("datafinal2.csv", encoding="utf-8", index=False)
```

```python
datafinal2 = datafinal2.round({"porodnost":2,"str_delka_ziv-CLKM":2,"str_delka_ziv-M":2,"str_delka_ziv-Z":2,"umrtnost":2,"mig_saldo_na_1000":2,"mig_saldo":2,"procentualni_narust_pop":2,"prirozeny_prirustek_na_1000":2})
datafinal2.head()
# datafinal2.info()
datafinal2.to_csv("datafinal2.csv", encoding="utf-8", index=False)
```
