```python
isocodes = pd.read_csv("iso_codes-countries.csv", sep=";", encoding="utf-8")
latlng = pd.read_csv("lat_lng-iso_2.csv", sep=";", encoding="utf-8")
```

```python
print(isolatlng)
```

```python
datamerge = pd.merge(databeta, isolatlng, left_on="LocID", right_on="id")
```
